## foreach

Im ersten Beispiel haben wir den gesamten DEAS-Datensatz vor uns und wollen ihn mit Hilfe von Schlagworten durchsuchen. Was lsst sich zum Thema *Gesundheit* finden? danach wollen wir es tabulieren. Zunchst einmal ohne Schleife, dann noch einmal mit:

```stata
lookfor Gesundheit
tab1 `r(varlist)'
```

Jetzt noch einmal tab1 als Schleife fr die letzten beiden Werte: 

```stata
lookfor Check
foreach var of varlist `r(varlist)'  {
    tab `var'
}
```

Der Name var ist beliebig whlbar und knnte zum Beispiel durch *Blume* ersetzt werden. 

```stata
lookfor Check
foreach Blume of varlist `r(varlist)'  {
    tab `Blume'
}
```

*varlist* gibt an worber die Schleife geht, hier ber Variablennamen. Und `' sind der Zugriff auf "locals" die nach vielen Befehlen von STATA verfgbar sind, z.B. nach lookfor (help lookfor), sind es nur die aufgelisteten Variablen (s. ganz unten):


Title

    [D] lookfor -- Search for string in variable names and labels


Syntax

        lookfor string [string [...]]


Description

    lookfor helps you find variables by searching for string among all
    variable names and labels.  If multiple strings are specified, lookfor
    will search for each of them separately.  You may search for a phrase by
    enclosing string in double quotes.


Examples

    ----------------------------------------------------------------------------
    Setup
        . webuse nlswork

    Find all occurrences of code in variable names and labels
        . lookfor code

    Find all occurrences of married in variable names and labels
        . lookfor married

    Find all occurrences of gnp in variable names and labels
        . lookfor gnp

    Find all occurrences of code or married in variable names and labels
        . lookfor code married

    ----------------------------------------------------------------------------
    Setup
        . sysuse sp500

    Find all occurrences of price in variable names and labels
        . lookfor price

    Find all occurrences of the phrase price change in variable names and
    labels
        . lookfor "price change"
    ----------------------------------------------------------------------------


Stored results

    lookfor stores the following in r():

    Macros    
      r(varlist)     the varlist of found variables


Daher: 

```stata
lookfor Check
foreach Blume of varlist `r(varlist)'  {
    tab `Blume'
}
```
